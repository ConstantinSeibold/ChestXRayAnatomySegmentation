```python
import torch
import numpy as np
from PIL import Image
import shutil
import os
os.chdir('../')
from cxas.segmentor import CXAS
import pandas as pd
```

```python

model = CXAS(
    model_name = 'UNet_ResNet50_default',
    gpus       = '0'
)
```

```python
input_path = './images'
out_path = './out_segs/'


print('Store segmentations for all files in folder as .npy-file')
print('Does the out_path exists:', os.path.isdir(out_path))

model.process_folder(
    input_path, 
    out_path, 
    'npy', 
    create=True
    )

print('First five Files in directory:', [os.path.join(dp, f) 
                              for dp, dn, fn in os.walk(os.path.expanduser(out_path)) for f in fn][:5])
shutil.rmtree(out_path)

print('Store segmentations for all files in folder as .npz-file')
print('Does the out_path exists:', os.path.isdir(out_path))

model.process_folder(
    input_path, 
    out_path, 
    'npz', 
    create=True
    )

print('First five Files in directory:', [os.path.join(dp, f) 
                              for dp, dn, fn in os.walk(os.path.expanduser(out_path)) for f in fn][:5])
shutil.rmtree(out_path)

print('Store segmentations for all files in folder as .png-file')
print('Does the out_path exists:', os.path.isdir(out_path))

model.process_folder(
    input_path, 
    out_path, 
    'png', 
    create=True
    )

print('First five Files in directory:', [os.path.join(dp, f) 
                              for dp, dn, fn in os.walk(os.path.expanduser(out_path)) for f in fn][:5])
shutil.rmtree(out_path)

print('Store segmentations for all files in folder as .jpg-file')
print('Does the out_path exists:', os.path.isdir(out_path))

model.process_folder(
    input_path, 
    out_path, 
    'jpg', 
    create=True
    )

print('First five Files in directory:', [os.path.join(dp, f) 
                              for dp, dn, fn in os.walk(os.path.expanduser(out_path)) for f in fn][:5])
shutil.rmtree(out_path)


print('Store segmentations for all files in folder as .jpg-file')
print('Does the out_path exists:', os.path.isdir(out_path))

model.process_folder(
    input_path, 
    out_path, 
    'json', 
    create=True
    )

print('First five Files in directory:', [os.path.join(dp, f) 
                              for dp, dn, fn in os.walk(os.path.expanduser(out_path)) for f in fn][:5])
shutil.rmtree(out_path)

input_path = './images/dcm_images'

print('Store segmentations for each segmentation label for all files in folder as dicom-seg (.dcm) in folder \n \
        This is only possible if all files in directory are dicom files!')
print('Does the out_path exists:', os.path.isdir(out_path))

model.process_folder(
    input_path, 
    out_path, 
    'dicom-seg', 
    create=True
    )

print('First five Files in directory:', [os.path.join(dp, f) 
                              for dp, dn, fn in os.walk(os.path.expanduser(out_path)) for f in fn][:5])
shutil.rmtree(out_path)
```
